```python
import numpy as np
import os

matrix = [
    [896.19990419, 0.0,          306.624076],
    [0.0,          896.80971174, 296.38767718],
    [0.0,          0.0,          1.0]
]

distortion = [[
    -1.52545659e-02,
     1.88969286e-01,
     3.26782932e-04,
    -1.96025757e-04,
    -3.67892820e-01
]]

save_path = os.path.expanduser("~/mycobot_ws/jetcobotcamera1.npz")

np.savez(
    save_path,
    mtx=np.array(matrix, dtype=np.float64),
    dist=np.array(distortion, dtype=np.float64)
)

print("Saved to:", save_path)
```

```python
data = np.load(save_path)
print("keys:", data.files)
print("camera_matrix:\n", data["mtx"])
print("dist_coeffs:\n", data["dist"])
```
